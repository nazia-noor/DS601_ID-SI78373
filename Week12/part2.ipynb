```python
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
ages = defaultdict(int)
occupations = defaultdict(int)

with open('data_big.csv') as file:
    c = 0
    while True:
        c += 1  
        line = file.readline()
        if not line: break
        elm = line.split(",")
        ages[int(elm[3])] += 1
        occupations[elm[4]] += 1
        if c%20000 == 0: print("Processed line", c)
print("Processed Done!")
```

```python
ages = dict(sorted(ages.items()))
```

A histogram of ages in logarithmic scale 

```python
fig = plt.figure(figsize=(25,6))
plt.bar([str(v) for v in ages.keys()], ages.values(), color='g')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.yscale('log')
plt.setp(plt.gca().get_xticklabels(), rotation=80, fontsize=12)
plt.setp(plt.gca().get_yticklabels(), fontsize=12)
plt.title('Histogram of ages in logarithmic scale',fontsize=16)
plt.show()
```
